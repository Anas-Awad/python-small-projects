```python
class User:
    def __init__(self,name,age,gender,ID,password):
        self.name = name
        self.gender = gender
        self.age =age
        self.ID = ID
        self.password = password
    def show_user_data(self):
        print(f'User Details:\nName : {self.name}\nAge  : {self.age}\nGender : {self.gender}')
```

```python
class Bank(User):
    
    def __init__(self,name,age,gender,ID,password):
        User.__init__(self,name,age,gender,ID,password)
        self.__balance = 0
        
        
    def deposit(self,amount):
        while True:
            passw = input('Password : ')
            if self.password == passw:
                self.amount = amount
                self.__balance += self.amount if self.amount > 0 and type(self.amount) == int else 0
                print('\n\n{}'.format('*'*25))
                print('wrong operation make sure you entered a right amount to deposit' if self.amount < 0 or type(self.amount) != int else f'operation succeeded!\nyour new balance is : {self.__balance} EGP')
                break
            else:
                print('Wrong password, Enter A Valid Password')
        
    def withdrawal(self,cashback):
        while True:
            passw = input('Password : ')
            if self.password == passw:
                self.cashback = cashback
                if self.cashback < 0 or type(self.cashback) != int or self.cashback > self.__balance:
                    print('\n\n{}'.format('*'*25))
                    print (f"Insufficient Funds: Balance Availble = {self.__balance} EGP" )
                else:
                    self.__balance -= self.cashback
                    print('\n\n{}'.format('*'*25))
                    print ("Operation Done succsessfuly")
                    print (f"Balance has been updated : {self.__balance} EGP ")
                break
            else:
                print('Wrong password, Enter A Valid Password')    
            
    def view_balance(self):
        while True:
            passw = input('Password : ')
            if self.password == passw:
                print('\n\n{}'.format('*'*25))
                print('Balance Availble = ', self.__balance)
                break
            else:
                print('Wrong password, Enter A Valid Password')
        
    def view_all_account_details(self):
        while True:
            passw = input('Password : ')
            if self.password == passw:
                print('\n\n{}'.format('*'*25))
                print(f'User Details:\nName : {self.name}\nAge  : {self.age}\nGender : {self.gender}\nAccount Balance : {self.__balance}')
                break
            else:
                print('Wrong password, Enter A Valid Password')
```

```python
def askint(word):
    while True:
        try:
            val = int(input(word))
            return val
        except:
            print('invalid input please try again'.title())
            
def ask_string(word):
    while True:
        flag = True
        val = input(word)
        for i in val:
            if not i.isalpha() and i != ' ' :
                flag = False
        if flag == True:
            return val
        else:
            print('invalid input please try again'.title())
def ask_id():
    while True:
        val = input('ID : ')
        flag = True
        for i in val:
            if not i.isalpha() and not i.isdigit() and i != ' ':
                flag = False
        if flag == True:
            return val
        else:
            print('ID Must Contain Only Letters And Digits, Try Again ^_^')
            continue
def ask_age():
    while True:
        try:
            val = int(input('Age : '))
            if val >= 10 and val <= 100:
                return val
            else:
                print('Enter A Normal Age Please!')
        except:
            print('Enter Integer Number ^_^')
def Gender():
    while True:
        try:
            val = int(input('Gender :\nEnter\n1 --> female\n2 --> male\n'))
            if val == 1:
                return 'Female'
            elif val == 2:
                return 'Male'
            else:
                print('Wrong Choice Please ReEnter Your Gender')
        except:
            print('Please Enter A Digit')
    
```

```python
def main():
    exit = False
    while not exit:
        try:
            print('{}'.format('*'*25))
            print('Welcome To Anas\'s Bank ^_^')
            user = Bank(ask_string('Name : '),ask_age(),Gender(),ask_id(),input('Password : '))
            print('Account Created Successfully!')
            while True:
                print('*'*25)
                val = askint('Enter\n1 to show all details account\n2 to show account balance\n3 to withdraw\n4 to deposit\n5 to Signout\n0 to Exit\n')
                if val == 5:
                    print('{}'.format('*'*25))
                    print(f'Good Bye {user.name} :\'(')
                    print('{}\n'.format('*'*25))
                    break
                elif val == 0:
                    print('{}\n'.format('*'*25))
                    print('Good Bye :\'(')
                    exit = True
                    break
                elif val == 1:
                    user.view_all_account_details()
                elif val == 2:
                    user.view_balance()
                elif val == 3:
                    amount = askint('Amount in EGP : ')
                    user.withdrawal(amount)
                elif val == 4:
                     amount = askint('Amount in EGP : ')
                     user.deposit(amount)
        except Exception as e:
               print(type(e).__name__)
```

```python
main()
```
